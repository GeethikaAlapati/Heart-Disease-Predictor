### SPLITTING FEATURES AND TARGET ###

```python
x=df.drop(columns='target',axis=1)
x
```

```python
y=df['target']
y
```

### TRAINING AND TESTING ###

```python
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)
```

### MODEL BUILDING 

Logistic Regression is used to create the model where end of it is binary classification.

```python
model=LogisticRegression()
```

```python
# Train the model

model.fit(x_train,y_train)
```

### EVALUATION

```python
# accuracy score on training data

x_train_prediction=model.predict(x_train)
training_accuracy=accuracy_score(x_train_prediction, y_train)
print(f"The training accuracy is {training_accuracy}")
```

```python
# accuracy score on testing data

x_test_prediction=model.predict(x_test)
testing_accuracy=accuracy_score(x_test_prediction, y_test)
print(f"The testing accuracy is {testing_accuracy}")
```

### BUILDING THE PREDICTING SYSTEM

```python
# One of the record from feature data as my input
input_data=(44,1,1,120,263,0,1,173,0,0,2,0,3)

# Converting it into numpy array
input_data=np.asarray(input_data)

# Reshaping into required input form
input_data=input_data.reshape(1,-1)

# With the input Iam predicting the output
prediction=model.predict(input_data)

if prediction[0]==0:
    print("Good News,The patient doen't have any heart disease")
else:
    print("The patient should consult a doctor")
```
